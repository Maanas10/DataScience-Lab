```python
# importing matplotlib module
from matplotlib import pyplot as plt
from matplotlib import style
 
style.use('ggplot')
# x-axis values
x = [5, 2 , 6 , 9, 4, 7 , 4]
# Y-axis values
y = [10, 5, 9 , 8, 4, 2 , 4 ]
# Function to plot
plt.plot(x,y)
plt.title('Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
# function to show the plot
plt.show()
```
