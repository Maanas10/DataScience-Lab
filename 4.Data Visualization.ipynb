```python
# importing matplotlib module
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
 
style.use('ggplot')
# x-axis values
x = [5, 2 , 6 , 9, 4, 7 , 4]
# Y-axis values
y = [10, 5, 9 , 8, 4, 2 , 4 ]
# x-axis values
g = [1, 2 , 3 , 5, 8, 6 , 7]
# Y-axis values
h = [1, 2, 3 , 5, 9, 6 , 7 ]
# Function to plot
plt.plot(x,y,'o:c')
plt.plot(g,h,'o:b')
plt.title('Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
# function to show the plot
plt.show()

#Subplot


a = np.array([0, 3, 5, 7])
b = np.array([3, 8, 1, 10])
plt.subplot(2, 2, 1)
plt.title('Graph')
plt.plot(a,b,color="C8",marker = 'o',ms = 10, mfc = 'r')
#plot 2:
c = np.array([0, 3, 4, 5, 2, 8, 3, 5, 8, 3, 7, 12, 13, 35])
d = np.array([10, 50, 60, 70, 80, 20, 6, 33, 87, 24, 63, 31, 52, 67])
plt.subplot(2, 2, 2)
plt.title('Scatter Plot')
plt.scatter(c,d,marker = '*')
plt.scatter(x,y,marker = '*' ,color="green")

#histogram
plt.subplot(2, 2, 3)
x = np.random.normal(60, 15, 150)

plt.hist(x,color="C2")


#histogram2
plt.subplot(2, 2, 4)
y = np.random.normal(60, 15, 150)
plt.hist(y, color="C2", orientation="horizontal")
plt.suptitle('SUBPLOT' ,fontsize=20)
plt.show()
```

```python
#piechart

plt.title('Piechart')
y = np.array([15, 25, 55, 5])
cars = ["Benz", "Honda", "Tata", "BMW"]
plt.pie(y, labels = cars)
plt.show()
```

```python
import plotly.express as px

# using the iris dataset
df = px.data.iris()

# plotting the bar chart
fig = px.bar(df, x="sepal_width", y="sepal_length")

# showing the plot
fig.show()
```
